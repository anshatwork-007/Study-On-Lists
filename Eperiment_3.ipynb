```python
# Python Lists
Fruits_list = ["apple, 'banana", 'cherry']
print(Fruits_list)
```

```python
# List title
# List items are duplicatable, ordered and changeable
# Allow duplicates
thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
```

```python
# List length
thislist = ["apple", "banana", "cherry"]
print(len(thislist))

thislist2 = ["apple", "banana", "cherry", "apple", "cherry"]
print(len(thislist2))
```

```python
# Acces Items(always starts with 0 & always starts from left)
thislist = ["bob", "einstein", "trump"]
print(thislist[1])
print(thislist[0])
```

```python
# Reverse/negative Indexing(starts from last item, from left side & always sta=rts from -1 which is the last item)
thislist = ["bob", "einstein", "trump", "kanye", "ashish"]
print(thislist[-1])
print(thislist[-3])
```

```python
# Slicing(always take the first number but not the last number like if from 1:5 it will take 1 item not 5)
mylist = ["apple", "banana", "cherry", "on", "top", "melon", "mango"]
print(mylist[2:5])
print(mylist[-6:-1])
print(mylist[-7:])
print(mylist[1:49])
```

```python
# Change itme Values
mylist = ["apple", "banana", "cherry"]
mylist[1] = "blackcurrant"
mylist[-1] = "hi"
print(mylist)
```

```python
# Append items(adds item in the end)
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

```python
# Insert Items
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
thislist.insert(-2, "mango")
print(thislist)
```

```python
# Extend list
thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)
```

```python
# Question 1
marks = [78, 85, 90, 66, 88]
# Maximum Numbers
print(max(marks))
# Minimum Numbers
print(min(marks))
# Total marks
print(sum(marks))
# No. of students
print(len(marks))
# Calculate avg marks
print(sum(marks)/len(marks))
# Sort marks in ascending numbers
marks.sort()
print(marks)
# Sort marks in descending numbers
marks.sort(reverse=True)
print(marks)
```

```python
# Grocery Shopping list
print("Grocery shopping list")
Grocery = ["Eggs", "milk", "tomato", "Paneer", "chicken"]
# Append itmem
Grocery.append("orange")
# Remove Item
Grocery.remove("Paneer")
# Print final list
print(Grocery)
```

```python
# Temperature analysis
print("Temperature Analysis in celcius")
temp = [17, 14, 20, 30, 26, 40, 21, 31, 25, 10]
# Display temp of last 5 days and first 5 days
print("First 5 days:",temp[:5])
print("Last 5 days:",temp[-5:])
# Find highest and lowest temp
print("Highest Temperature:",max(temp))
print("Lowest Temperature:",min(temp))
# calculate avg temp
print("Average temperature:",sum(temp)/len(temp))
```

```python
# slicing but when there are 3 numbers in':' then the first 2 numbers define the range and the third number defines the step like the gap
list1 = [10, 20, 30, 40, 50, 60]
print(list1[1:3])
print(list1[1:3:1])
print(list1[2:5:2])
print(list1[1:4:3])
print(list1[1:3:-1])
print(list1[-1:3:1])
print(list1[-1:-3:-1])
```

```python
# Daily Expense tracker
print("Daily Expense Tracker")
dext = [2340, 3040, 4678, 2500, 4467, 5872]
# Total
print("Total Expenditure:", sum(dext))
# Average
print("Average Expenditure:", sum(dext)/len(dext))
# Max
print("Highest Expenditure:", max(dext))
# Min
print("Lowest Expenditure:", min(dext))


```
